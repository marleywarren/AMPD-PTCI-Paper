```python
# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/marleywarren/Dropbox (Rush)/Projects/AMPD+PTCI/Mech Data Straight from RedCap.csv', low_memory=False, encoding='latin1')
df.rename(columns={'ï»¿part_id': 'part_id'}, inplace=True)

# Set the multi-index with time and name as the index
df.set_index(['redcap_event_name', 'part_id'], inplace=True)

# Unstack the time index to create separate columns for each unique time
df = df.unstack(level=0)

# Flatten the column index to have a single level of column names
df.columns = [f'{col[1]}_{col[0]}' for col in df.columns]

df = df.dropna(axis=1, how='all')
#df = df.dropna(subset=['arm_1_intake_pcl5_score_pastmonth'])

# Combine like columns from different arms into one column using np.where
arm_2_columns = [col for col in df.columns if "_arm_2_" in col]

for arm_2_col in arm_2_columns:
    beg, end = arm_2_col.split("_arm_2_")
    arm_3_col = beg + "_arm_3_" + end
    arm_23_col = beg + "_arm_23_" + end
    
    if arm_3_col not in df.columns:
        df = df.drop(columns=[arm_2_col])
        continue
    
    df[arm_23_col] = np.where(df[arm_2_col].isnull(), df[arm_3_col], df[arm_2_col])
    df = df.drop(columns=[arm_2_col, arm_3_col])

df.to_csv('Angie_MPA_data_unstacked.csv')
```
